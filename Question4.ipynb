For a List A, B, C, D, E write a python program to compute all the combinations and permutations 

```python
import itertools

# Define the list
elements = ["A", "B", "C", "D", "E"]

# Compute all permutations
def permutations(elements):
    print("Permutations: ")
    permutations = list(itertools.permutations(elements))
    for perm in permutations:
        print(perm)

# Compute all combinations of length 2, 3, 4, and 5
def combinations(elements):
    print("\nCombinations: ")
    for r in range(1, len(elements) + 1):
        combinations = list(itertools.combinations(elements, r))
        print(f"Combinations of length {r}:")
        for comb in combinations:
            print(comb)

import cProfile
def profile_operations():
    permutations(elements)
    combinations(elements)
    print("\n\n Performing profiling")

cProfile.run('profile_operations()')
```
